```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Vehicle_Information.csv'  # Update with the path to your dataset

# Attempt to read the CSV file with a specified encoding
try:
    df = pd.read_csv(file_path, encoding='ISO-8859-1')  # Try 'ISO-8859-1' if you encounter encoding issues
except UnicodeDecodeError:
    # If an error occurs, try a different encoding
    df = pd.read_csv(file_path, encoding='windows-1252')

# Clean up the dataset (optional step to handle NaN values or other cleaning tasks)

# Visualization

# Accidents by Driver's Home Area Type
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Driver_Home_Area_Type', palette='viridis')
plt.title('Accidents by Driver\'s Home Area Type')
plt.xlabel('Driver\'s Home Area Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Accidents by Age Band of Driver
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Age_Band_of_Driver', palette='viridis')
plt.title('Accidents by Age Band of Driver')
plt.xlabel('Age Band of Driver')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Accidents by Year
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Year', palette='viridis')
plt.title('Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Accidents by Vehicle Type
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Vehicle_Type', palette='viridis')
plt.title('Accidents by Vehicle Type')
plt.xlabel('Number of Accidents')
plt.ylabel('Vehicle Type')
plt.tight_layout()
plt.show()

# Accidents by Junction Location
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Junction_Location', palette='viridis')
plt.title('Accidents by Junction Location')
plt.xlabel('Number of Accidents')
plt.ylabel('Junction Location')
plt.tight_layout()
plt.show()

# Correlation Matrix (for numeric features only)
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
plt.figure(figsize=(14, 12))
correlation_matrix = numeric_df.corr()  # Compute correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.show()
```
